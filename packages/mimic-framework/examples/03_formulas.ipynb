```python
from mimic.formulas import (
    dcf_impact,
    cogs_sensitivity,
    inventory_burn,
    fx_passthrough,
    supplier_hhi,
    cascade_propagate,
    bayes_update,
    capm_response,
    operating_leverage,
    altman_z,
)
```

```python
# 1. How much does a 15% input cost spike hurt WMT's margin?
result = cogs_sensitivity(
    revenue=648_125,   # $M
    cogs=487_988,
    input_shock_pct=0.15,
    passthrough_rate=0.40,   # WMT can pass 40% of cost increases
)
print("COGS Sensitivity:")
print(f"  Base gross margin: {result['base_gross_margin']:.1%}")
print(f"  New gross margin:  {result['new_gross_margin']:.1%}")
print(f"  Margin compression: {result['margin_compression']:.1%}")
print(f"  EBITDA impact: ${result['annual_ebitda_impact_usdM']:,.0f}M")
```

```python
# 2. Inventory burn — how many days of buffer if demand spikes 50%?
result = inventory_burn(inventory_days=42.3, demand_multiplier=1.5)
print("Inventory Burn:")
print(f"  Effective days remaining: {result['effective_days_remaining']:.1f}")
print(f"  Risk level: {result['risk_level']}")
print(f"  Restocking urgency: {result['restocking_urgency']}")
```

```python
# 3. FX impact — EUR/GBP/CAD weaken 5% vs USD
result = fx_passthrough(
    revenue=648_125,
    fx_exposure={"EUR": 0.08, "GBP": 0.05, "CAD": 0.06},
    fx_moves={"EUR": -0.05, "GBP": -0.05, "CAD": -0.05},
)
print("FX Passthrough:")
print(f"  Total revenue impact: ${result['total_revenue_impact_usdM']:,.0f}M")
print(f"  As % of revenue: {result['pct_of_revenue']:.2%}")
print(f"  By currency: {result['breakdown_by_currency']}")
```

```python
# 4. Supplier concentration risk
result = supplier_hhi({"China": 30, "USA": 50, "Vietnam": 20})
print("Supplier HHI:", result)
```

```python
# 5. Cascade propagation of a port closure
result = cascade_propagate(initial_shock=0.75, n_tiers=3, amplification_risk=0.25)
print("Cascade Propagation:", result)
```

```python
# 6. Bayes update — how does news of a port closure change crisis probability?
result = bayes_update(
    prior=0.20,               # 20% chance of severe disruption before news
    likelihood_given_true=0.85,   # if disruption is coming, 85% chance we'd see this news
    likelihood_given_false=0.10,  # if not, only 10% chance
)
print("Bayesian Update:")
print(f"  Prior: {result['prior']:.0%}  →  Posterior: {result['posterior']:.0%}")
print(f"  Direction: {result['direction']}")
```
