# mimic-bench: Running the Benchmark

This notebook shows how to evaluate any predict function against the labeled dataset.

```python
import anthropic
from mimic_bench import Benchmark

client = anthropic.Anthropic()

def claude_predict(prompt: str) -> str:
    msg = client.messages.create(
        model="claude-opus-4-7",
        max_tokens=512,
        messages=[{"role": "user", "content": prompt}],
    )
    return msg.content[0].text
```

```python
bench = Benchmark.load("v1", predict_fn=claude_predict)

# Run on a single event first
results = bench.run(
    event_ids=["2020_03_covid_lockdowns"],
    verbose=True,
    delay=0.5,
)
print(results)
```
