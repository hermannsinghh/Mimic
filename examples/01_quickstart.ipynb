# Mimic Quickstart

Build a digital twin from any ticker and simulate how it responds to real-world events.

```python
import os
os.environ["OPENAI_API_KEY"] = "sk-..."  # or load from .env
```

```python
from mimic import Twin

# Pull from SEC EDGAR + yfinance + 10-K automatically
twin = Twin.from_ticker("WMT")
print(twin)
```

```python
# Show the full company profile
print(twin.context.summary())
```

```python
# Simulate an event
result = twin.simulate(
    event="China port closes for 30 days",
    severity=0.75,
)
print(result.pretty())
```

```python
# Access structured fields directly
print(f"Immediate action: {result.immediate_action_0_24h}")
print(f"P50 financial impact: ${result.financial_impact_mid:,.0f}M")
print(f"Confidence: {result.confidence:.0%}")
print(f"Secondary risks: {result.secondary_risks_created}")
```
