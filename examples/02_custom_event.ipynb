# Custom Events & Multiple Tickers

Compare how different companies respond to the same macro event.

```python
import os
os.environ["OPENAI_API_KEY"] = "sk-..."
```

```python
from mimic import Twin

EVENT = "Fed raises rates 100bps unexpectedly"
TICKERS = ["JPM", "WMT", "AAPL"]

results = {}
for ticker in TICKERS:
    twin = Twin.from_ticker(ticker)
    results[ticker] = twin.simulate(EVENT, severity=0.65)
```

```python
# Compare P50 financial impact across companies
import pandas as pd

rows = [
    {
        "Ticker": ticker,
        "P50 Impact ($M)": r.financial_impact_mid,
        "Confidence": f"{r.confidence:.0%}",
        "Immediate Action": r.immediate_action_0_24h[:80],
    }
    for ticker, r in results.items()
]

pd.DataFrame(rows)
```
